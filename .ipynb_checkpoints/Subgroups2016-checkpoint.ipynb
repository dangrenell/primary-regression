```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
import warnings
# warnings.filterwarnings("ignore")
```

```python
demographics = pd.read_csv('county_facts.csv')
demographics = demographics[['fips','area_name','state_abbreviation','PST045214','PST120214',
                             'AGE775214','SEX255214','RHI225214','RHI725214','RHI825214',
                             'POP645213','POP815213','EDU635213','EDU685213','INC110213',
                             'PVY020213','POP060210']]
columns = {'PST045214':'Population',
           'PST120214':'Population Change',
           'AGE775214':'Population Over 65',
           'SEX255214':'Percent Female',
           'RHI225214':'Percent Black',
           'RHI725214':'Percent Latino',
           'RHI825214':'Percent White',
           'POP645213':'Percent Foreign Born',
           'POP815213':'Percent LOTE',
           'EDU635213':'HighSchool',
           'EDU685213':'Bachelors',
           'INC110213':'Median Household Income',
           'PVY020213':'Percent in Poverty',
           'POP060210':'Population Per Sq. Mi.'}

demographics.rename(columns = columns, inplace=True)


results = pd.read_csv('primary_results.csv')
results = results[results.party == "Democrat"]
results = results[(results.candidate == "Hillary Clinton") |(results.candidate == "Bernie Sanders") ]
Dem = results
```

```python
graph1 = Dem.groupby(['candidate'])['votes'].sum()
graph1.plot.bar()
graph1
```

```python
clinton_state_totals = Dem[Dem['candidate']=='Hillary Clinton'].groupby('state')['votes'].sum()
sanders_state_totals = Dem[Dem['candidate']=='Bernie Sanders'].groupby('state')['votes'].sum()
difference = (clinton_state_totals-sanders_state_totals)/(clinton_state_totals+sanders_state_totals)
difference.sort_values()
```

```python
graph1 = Dem[Dem['state']=='Vermont'].groupby(['candidate'])['votes'].sum()
graph1.plot.bar()
graph1
```

```python
graph2 = Dem[Dem['state']=='Mississippi'].groupby(['candidate'])['votes'].sum()
graph2.plot.bar()
graph2
```

```python
votesByState = [[candidate, state, party] for candidate in Dem.candidate.unique() for state in Dem.state.unique() for party in Dem.party.unique()]
for item in votesByState:
    item.append(Dem[(Dem.candidate == item[0]) & (Dem.state == item[1])].votes.sum())
    item.append(item[3]*1.0/Dem[Dem.state == item[1]].votes.sum())
vbs = pd.DataFrame(votesByState, columns = ['candidate', 'state', 'party', 'votes','partyFrac'])
```

```python
allData = pd.merge(vbs, demographics, how="inner", left_on = 'state',right_on = 'area_name')
allData.drop('state_abbreviation',axis=1, inplace=True)

Clinton = allData[(allData.candidate == "Hillary Clinton")]
Clinton = Clinton.reset_index();
Clinton.drop('index',axis=1, inplace=True)


random_states = (Clinton.state == 'Oklahoma')|(Clinton.state == 'Massachusetts')|(Clinton.state == 'Rhode Island')|(Clinton.state == 'Vermont')|(Clinton.state == 'Kentucky')

feature_cols = list(columns.values())

X_train = Clinton[~random_states].drop(['partyFrac'], axis = 1)
X_train = X_train[feature_cols]
X_test = Clinton[random_states].drop(['partyFrac'], axis = 1)
X_test = X_test[feature_cols]
y_train = Clinton[~random_states].partyFrac
y_test = Clinton[random_states].partyFrac
```

```python
sandersData = allData[allData['candidate']=='Bernie Sanders']
sandersData = sandersData[sandersData['partyFrac']>=.5]

clintonData = allData[allData['candidate']=='Hillary Clinton']
clintonData = clintonData[clintonData['partyFrac']>=.5]

winningData = pd.concat([sandersData, clintonData])
```

```python
scatter_of_candidates = sns.scatterplot(x='Percent Black', y='HighSchool', data=winningData, hue='candidate')
```

```python
variable_comparison_graph = sns.pairplot(winningData, vars=['votes', 'Population',
       'Percent Female', 'Percent Black', 'Percent Latino', 'Percent White',
       'Percent LOTE', 'HighSchool', 'Bachelors', 'Percent in Poverty',
       'Population Per Sq. Mi.'], hue='candidate')
```

```python
allCountyData = pd.merge(Dem, demographics, how="inner", left_on = 'fips',right_on = 'fips')

sandersCountyData = allCountyData[allCountyData['candidate']=='Bernie Sanders']
sandersCountyData = sandersCountyData[sandersCountyData['fraction_votes']>=.5]

clintonCountyData = allCountyData[allCountyData['candidate']=='Hillary Clinton']
clintonCountyData = clintonCountyData[clintonCountyData['fraction_votes']>=.5]

winningCountyData = pd.concat([sandersCountyData, clintonCountyData])
```

```python
clintonCountyData.plot.scatter(x='Percent Black', y='HighSchool', c='#ff7f0e')
sandersCountyData.plot.scatter(x='Percent Black', y='HighSchool', c='#1f77b4')
```

```python
ClintonCounty = allCountyData[(allCountyData.candidate == "Hillary Clinton")]
ClintonCounty = ClintonCounty.reset_index();
ClintonCounty.drop('index',axis=1, inplace=True)

X = ClintonCounty.drop('fraction_votes', axis = 1)
y = ClintonCounty['fraction_votes']

X_trainCounty, X_testCounty, y_trainCounty, y_testCounty = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_trainCounty = X_trainCounty[feature_cols]
X_testCounty = X_testCounty[feature_cols]
```

```python
rfRegressorCounty = RandomForestRegressor(n_estimators=100)
adaRegressorCounty = AdaBoostRegressor(n_estimators=100)
gbRegressorCounty = GradientBoostingRegressor(n_estimators=100)

rfRegressorCounty.fit(X_trainCounty,y_trainCounty)
print(rfRegressorCounty.feature_importances_)

adaRegressorCounty.fit(X_trainCounty,y_trainCounty)
print(adaRegressorCounty.feature_importances_)

gbRegressorCounty.fit(X_trainCounty,y_trainCounty)
print(gbRegressorCounty.feature_importances_)
```

```python
y_rfpred = rfRegressorCounty.predict(X_testCounty)
print(mean_squared_error(y_testCounty, y_rfpred))

y_adapred = adaRegressorCounty.predict(X_testCounty)
print(mean_squared_error(y_testCounty, y_adapred))

y_gbpred = gbRegressorCounty.predict(X_testCounty)
print(mean_squared_error(y_testCounty, y_gbpred))
```

```python
y_rfpred = rfRegressor.predict(X_test)
print(mean_squared_error(y_test, y_rfpred))

y_adapred = adaRegressor.predict(X_test)
print(mean_squared_error(y_test, y_adapred))

y_gbpred = gbRegressor.predict(X_test)
print(mean_squared_error(y_test, y_gbpred))
```
